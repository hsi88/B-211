```
#a. Look at the distributions for each of the traits (sepal length, sepal width, petal length, and petal width). 

def his(column):
 plt.hist(iris[column], 30)
 plt.title('distirbution of ' + column, size = 20)
 plt.xlabel(column, size =20)
 plt.ylabel('frenquency',size=20)
 
 plt.savefig('distirbution of ' + column)
 plt.show()
 
for x in iris.columns[0:4]:
  his(x)



  
```

```

# b. Create a categorical plot for each trait by species
def categorical(tar):
  y=iris[iris.columns[tar]]
  rect = plt.bar(iris[iris.columns[4]], y)
  
#title, label
  plt.title("category for " + iris.columns[tar], size=20)
  plt.xlabel(iris.columns[tar],size=20)
  plt.ylabel('frenquency', size = 20)
  plt.savefig("category for " + iris.columns[tar], dpi=100)
#grid
  plt.minorticks_on() 
  plt.grid(which='major', linestyle='-', color='red')
  plt.grid(which='minor', linestyle=':', color='black')
 
  plt.show()

#iterate the column
for tar in range(4):
  categorical(tar)
 

```

```
#c. Create relational plots between trait pairs (sepal length vs sepal width, petal
# length vs petal width), include a way of distinguishing by species. 

sns.scatterplot(iris['sepal_length'], iris['petal_width'], hue=iris['species'])
plt.title('trait pairs (sepal length vs sepal width)') 
plt.savefig('trait pairs (sepal length vs sepal width)')

```

```
sns.scatterplot(iris['petal_length'], iris['petal_width'], hue=iris['species'])
plt.title('trait pairs (petal length vs petal width)')
plt.savefig('trait pairs (petal length vs petal width))')
```

```
import seaborn as sns
ex = sns.load_dataset('exercise')
ex
```

```
a. Create a heatmap for the participants, workout time, and pulse. (Hint use the pivot
command from Pandas to get the data in appropriate format) 


```

```
df = pd.pivot_table(data=sns.load_dataset("exercise"),
                    index='id',
                    values='pulse',
                    columns='time')


df
```

```

plt.figure(figsize = (16,8))

sns.heatmap(df,annot=True,  cmap="YlGnBu") 
plt.title('heatmap of pulse ')
plt.savefig('heatmap of pulse ')
```

```
b. Create a categorical plot for pulse by diet and by type of exercise. 
```

```

plt.figure(figsize = (16,8))
sns.stripplot(data=ex,x="diet", y="pulse")
stitle = 'pulse of different diet'
plt.title('pulse of different diet')
plt.savefig(stitle)
```

```

plt.figure(figsize = (16,8))
sns.stripplot(data=ex,x="kind", y="pulse")
plt.title('pulse of different exercise')
plt.savefig('pulse of different exercise')
```
